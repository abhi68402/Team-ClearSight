```python
import os
import cv2
import random
from glob import glob

base_dir = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/low_res_photos/dataset"

lr_dir = os.path.join(base_dir, "lr1")
hr_dir = os.path.join(base_dir, "hr1")

os.makedirs(lr_dir, exist_ok=True)
os.makedirs(hr_dir, exist_ok=True)

image_paths = sorted(glob(os.path.join(base_dir, "*.png")))

upscale_factors = [1.5, 1.7]

def bilateral_downscale(img, scale_factor):
    height, width = img.shape[:2]
    new_size = (int(width * scale_factor), int(height * scale_factor))
    upscaled = cv2.resize(img, new_size, interpolation=cv2.INTER_CUBIC)
    filtered = cv2.bilateralFilter(upscaled, d=9, sigmaColor=75, sigmaSpace=75)
    downscaled = cv2.resize(filtered, (width, height), interpolation=cv2.INTER_AREA)
    return downscaled

for idx, img_path in enumerate(image_paths[:2000], 1):
    img = cv2.imread(img_path)
    if img is None:
        print(f"Warning: Could not read {img_path}")
        continue

    scale = random.choice(upscale_factors)
    method = random.choice(['bicubic', 'bilateral'])

    height, width = img.shape[:2]
    hr_size = (int(width * scale), int(height * scale))
    hr_img = cv2.resize(img, hr_size, interpolation=cv2.INTER_CUBIC)

    if method == 'bicubic':
        lr_img = cv2.resize(hr_img, (width, height), interpolation=cv2.INTER_CUBIC)
    else:
        lr_img = bilateral_downscale(hr_img, 1/scale)

    fname = f"{idx:04d}.png"
    cv2.imwrite(os.path.join(hr_dir, fname), hr_img)
    cv2.imwrite(os.path.join(lr_dir, fname), lr_img)

    if idx % 100 == 0:
        print(f"Processed {idx} images")

print("Finished processing 2000 images into hr1 and lr1 inside the dataset folder.")
```

```python
import os
import cv2
import random
from glob import glob

base_dir = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/low_res_photos/dataset"

lr_dir = os.path.join(base_dir, "LR1")
hr_dir = os.path.join(base_dir, "HR1")

os.makedirs(lr_dir, exist_ok=True)
os.makedirs(hr_dir, exist_ok=True)

image_paths = sorted(glob(os.path.join(base_dir, "*.jpg")))

upscale_factors = [1.5, 1.7]

def bilateral_downscale(img, scale_factor):
    height, width = img.shape[:2]
    new_size = (int(width * scale_factor), int(height * scale_factor))
    upscaled = cv2.resize(img, new_size, interpolation=cv2.INTER_CUBIC)
    filtered = cv2.bilateralFilter(upscaled, d=9, sigmaColor=75, sigmaSpace=75)
    downscaled = cv2.resize(filtered, (width, height), interpolation=cv2.INTER_AREA)
    return downscaled

for idx, img_path in enumerate(image_paths[:200], 1):
    img = cv2.imread(img_path)
    if img is None:
        print(f"Warning: Could not read {img_path}")
        continue

    scale = random.choice(upscale_factors)
    method = random.choice(['bicubic', 'bilateral'])

    height, width = img.shape[:2]
    hr_size = (int(width * scale), int(height * scale))
    hr_img = cv2.resize(img, hr_size, interpolation=cv2.INTER_CUBIC)

    if method == 'bicubic':
        lr_img = cv2.resize(hr_img, (width, height), interpolation=cv2.INTER_CUBIC)
    else:
        lr_img = bilateral_downscale(hr_img, 1/scale)

    fname = f"{idx:04d}.jpg"
    cv2.imwrite(os.path.join(hr_dir, fname), hr_img)
    cv2.imwrite(os.path.join(lr_dir, fname), lr_img)

    if idx % 20 == 0:
        print(f"Processed {idx} images")

print("Done: 200 images processed and saved in HR1 and LR1.")
```
