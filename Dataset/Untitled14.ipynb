```python
import os
from PIL import Image

base_path = "/content/unsplash_dataset/Image Super Resolution - Unsplash"

lr_source = os.path.join(base_path, "low res")
hr_source = os.path.join(base_path, "high res")

dest_root = os.path.join(base_path, "dataset_root2")
lr_dest = os.path.join(dest_root, "lr2")
hr_dest = os.path.join(dest_root, "hr2")

os.makedirs(lr_dest, exist_ok=True)
os.makedirs(hr_dest, exist_ok=True)

def resize_image(img, scale_factor, method):
    w, h = img.size
    if scale_factor > 1:
        new_size = (int(w / scale_factor), int(h / scale_factor))
    else:
        new_size = (int(w * (1/scale_factor)), int(h * (1/scale_factor)))

    if method == 'bicubic':
        resample_method = Image.BICUBIC
    elif method == 'bilinear':
        resample_method = Image.BILINEAR
    else:
        raise ValueError("Method must be 'bicubic' or 'bilinear'")

    return img.resize(new_size, resample=resample_method)

scale = 1.7
methods = ['bicubic', 'bilinear']

print("Processing HR images for downscale...")

for hr_filename in os.listdir(hr_source):
    hr_path = os.path.join(hr_source, hr_filename)
    try:
        hr_img = Image.open(hr_path).convert('RGB')
    except Exception as e:
        print(f"Error opening HR image {hr_filename}: {e}")
        continue

    hr_dest_path = os.path.join(hr_dest, hr_filename)
    hr_img.save(hr_dest_path)

    # Downscale HR image by 1.7x using bicubic and bilinear
    for method in methods:
        lr_img = resize_image(hr_img, scale, method)
        name, ext = os.path.splitext(hr_filename)
        lr_filename = f"{name}_{method}{ext}"
        lr_dest_path = os.path.join(lr_dest, lr_filename)
        lr_img.save(lr_dest_path)

print("Processing LR images for upscale...")

for lr_filename in os.listdir(lr_source):
    lr_path = os.path.join(lr_source, lr_filename)
    try:
        lr_img = Image.open(lr_path).convert('RGB')
    except Exception as e:
        print(f"Error opening LR image {lr_filename}: {e}")
        continue

    lr_dest_path = os.path.join(lr_dest, lr_filename)
    lr_img.save(lr_dest_path)

    for method in methods:
        hr_img = resize_image(lr_img, 1/scale, method)
        name, ext = os.path.splitext(lr_filename)
        hr_filename = f"{name}_{method}{ext}"
        hr_dest_path = os.path.join(hr_dest, hr_filename)
        hr_img.save(hr_dest_path)

print("Done! Check your dataset_root2 folder.")
```

```python
import os
import shutil
from PIL import Image

lr_path = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/dataset_root2/lr2"
hr_path = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/dataset_root2/hr2"


clean_lr = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/final_dataset/lr2"
clean_hr = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/final_dataset/hr2"
os.makedirs(clean_lr, exist_ok=True)
os.makedirs(clean_hr, exist_ok=True)

valid_count = 0
skipped = 0

for fname in os.listdir(lr_path):
    lr_file = os.path.join(lr_path, fname)
    hr_file = os.path.join(hr_path, fname)

    # Skip if pair doesn't exist
    if not os.path.exists(hr_file):
        skipped += 1
        continue

    try:
        lr_img = Image.open(lr_file)
        hr_img = Image.open(hr_file)
    except:
        skipped += 1
        continue

    lr_w, lr_h = lr_img.size
    hr_w, hr_h = hr_img.size

    scale_w = hr_w / lr_w
    scale_h = hr_h / lr_h

    if 1.6 <= scale_w <= 1.8 and 1.6 <= scale_h <= 1.8:
        shutil.copy2(lr_file, os.path.join(clean_lr, fname))
        shutil.copy2(hr_file, os.path.join(clean_hr, fname))
        valid_count += 1
    else:
        skipped += 1

print(f"‚úÖ {valid_count} valid LR-HR pairs copied to final_dataset/")
print(f"‚ùå {skipped} files skipped due to missing pair or scale mismatch.")
```

```python
import os
from PIL import Image

hr_source = "/content/unsplash_dataset/Image Super Resolution - Unsplash/high res"


lr_dest = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/final_dataset/lr2"
hr_dest = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/final_dataset/hr2"

os.makedirs(lr_dest, exist_ok=True)
os.makedirs(hr_dest, exist_ok=True)

scale = 1.7
method = Image.BICUBIC

valid_count = 0
skipped = 0

for fname in os.listdir(hr_source):
    if not fname.lower().endswith(('.jpg', '.png', '.jpeg')):
        continue
    hr_path = os.path.join(hr_source, fname)
    try:
        hr_img = Image.open(hr_path).convert('RGB')
        w, h = hr_img.size
        new_size = (int(w / scale), int(h / scale))
        lr_img = hr_img.resize(new_size, resample=method)


        hr_img.save(os.path.join(hr_dest, fname))
        lr_img.save(os.path.join(lr_dest, fname))
        valid_count += 1
    except:
        skipped += 1
        continue

print(f"‚úÖ {valid_count} aligned LR-HR pairs created at 1.7x scale")
print(f"‚ùå {skipped} files skipped due to errors")
```

```python
import os
from PIL import Image
import shutil

hr_source = "/content/unsplash_dataset/Image Super Resolution - Unsplash/high res"

lr_temp = "/content/dataset_root2/lr2"
hr_temp = "/content/dataset_root2/hr2"
os.makedirs(lr_temp, exist_ok=True)
os.makedirs(hr_temp, exist_ok=True)

scale = 1.7
count = 0
skipped = 0

for fname in os.listdir(hr_source):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    hr_path = os.path.join(hr_source, fname)
    try:
        hr_img = Image.open(hr_path).convert("RGB")
        w, h = hr_img.size
        lr_size = (int(w / scale), int(h / scale))

        lr_img = hr_img.resize(lr_size, resample=Image.BICUBIC)


        lr_img.save(os.path.join(lr_temp, fname))
        hr_img.save(os.path.join(hr_temp, fname))

        count += 1
    except Exception as e:
        print(f"‚ùå Skipped {fname} due to error: {e}")
        skipped += 1

print(f"‚úÖ Generated {count} valid LR-HR pairs (bicubic, 1.7x downscaled)")
print(f"‚ö†Ô∏è Skipped {skipped} files")

from google.colab import drive
drive.mount('/content/drive')

drive_base = "/content/drive/MyDrive/INTEL PROJECT/archive/DATASET/dataset_root2"
lr_drive = os.path.join(drive_base, "lr2")
hr_drive = os.path.join(drive_base, "hr2")

os.makedirs(lr_drive, exist_ok=True)
os.makedirs(hr_drive, exist_ok=True)

shutil.copytree(lr_temp, lr_drive, dirs_exist_ok=True)
shutil.copytree(hr_temp, hr_drive, dirs_exist_ok=True)

print("üöÄ Final dataset copied to Drive under dataset_root2/lr2 and hr2")
```
